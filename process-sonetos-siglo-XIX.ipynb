```python
base_url = 'http://www.cervantesvirtual.com/obra-visor/sonetos-del-siglo-xix--0/html/'
```

```python
def get_cervantes_virtual_urls(base_url):
    f = urllib.request.urlopen(base_url + "00199522-82b2-11df-acc7-002185ce6064.html") 
    soup = BeautifulSoup(f, 'html.parser')
    urls = []
    for p in soup.findAll("p", {"class": "links"}):
        a = p.find("a")
        urls.append(base_url + a['href'].split('#')[0])
    return list((set(urls)))

def get_cervantes_virtual_pages(unique_urls):
    pages = []
    for i, page in enumerate(unique_urls):
        f_page = urllib.request.urlopen(page)
        soup_page = BeautifulSoup(f_page, 'html.parser')
        pages.append(soup_page)
        print('#', end='')
    return pages

def scrap_cervantes_virtual_pages(pages):
    sonetos = []
    for index, page in enumerate(pages):
        table = page.find('table')
        while table is not None:
            soneto = ""
            versos = 0
            tds = table.find_all('td', style="white-space:nowrap;")
            for td in tds:
                versos += 1
                soneto = soneto + td.text.replace('\n',' ').replace('\xa0\xa0\xa0','').strip(' ') + '\n'
                if versos in [4, 8, 11, 14]:
                    soneto = soneto + '\n'
            table = table.find_next_sibling('table')
            if versos == 14:
                if '[. . . . . . . . . . . . . . . . . . . . . . . . . . . . . . . . . . . . ]' in soneto:
                    print('[. . . . . . . . . . . . . . . . . . . . . . . . . . . . . . . . . . . . ]')
                else:
                    sonetos.append(soneto)
            else:
                print(unique_urls[index])
                print(soneto[:40])
                print('error: versos->', versos)
                print()
        print('total =',len(sonetos))
    return sonetos
```

```python
unique_urls = get_cervantes_virtual_urls(base_url)
```

```python
pages = get_cervantes_virtual_pages(unique_urls)
```

```python
sonetos = scrap_cervantes_virtual_pages(pages)
```
