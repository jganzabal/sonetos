```python
import numpy as np
from keras.models import Sequential
from keras.optimizers import Adam
from keras.layers import Embedding, LSTM, Dense
```

```python
VOCAB_SIZE = 24679
word_dim = 64
lstm_units = 128
model = Sequential()
model.add(Embedding(VOCAB_SIZE+1, word_dim, mask_zero = True))
model.add(LSTM(lstm_units, return_sequences=True))
model.add(Dense(VOCAB_SIZE, activation="softmax"))
model.summary()
```

```python
model.load_weights('weightsSonetosSigloXIX-2000-rmsprop-0.001-random_generator-16.hdf5')
```

```python
acc_1 = np.load('acc_historySonetosSigloXIX-2000-rmsprop-0.001-random_generator-16.npy')
loss_1 = np.load('loss_historySonetosSigloXIX-2000-rmsprop-0.001-random_generator-16.npy')
acc_2 = np.load('acc_historySonetosSigloXIX-882-adam-0.001-random_generator-16.npy')
loss_2 = np.load('loss_historySonetosSigloXIX-882-adam-0.001-random_generator-16.npy')
```

```python
window = 35
from matplotlib import pyplot as plt
plt.plot(acc_1[:window])
plt.plot(acc_2[:window])
plt.show()
plt.plot(loss_1[:window])
plt.plot(loss_2[:window])
plt.show()
```
